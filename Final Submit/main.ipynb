```python
import pandas as pd

# Load the dataset without columns
file_path = r'C:\Users\clint\Desktop\EvaluativeAssignment_2024\EvaluativeAssignment_2024\Medicare_Advantage\Medicare_Advantage\scp-1205.csv'
data = pd.read_csv(file_path, header=None)
data
```

```python
# Assign column names
data.columns = [
    'countyname', 'state', 'healthplanname', 'typeofplan', 'countyssa', 
    'eligibles', 'enrollees', 'penetration', 'ABrate', 'monthly_payment'
]

# Display the first few rows of the dataset
data
```

```python
#investigate the data types of the columns
data.dtypes
```

```python
# Convert columns to numeric types
data[['eligibles', 'enrollees', 'penetration', 'ABrate', 'monthly_payment']] = data[['eligibles', 'enrollees', 'penetration', 'ABrate', 'monthly_payment']].apply(pd.to_numeric, errors='coerce')

# Display the data types to verify the changes
data.dtypes
```

```python
import pandas as pd

# check data type
columns_to_convert = ['countyname', 'state', 'healthplanname', 'typeofplan', 'countyssa']
data[columns_to_convert] = data[columns_to_convert].astype(str)
data.dtypes
```

```python
import pandas as pd

#remove bad state entries
valid_states = [
    'AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 
    'MD', 'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 
    'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY'
]
# Strip leading and trailing whitespaces from the state column and convert to uppercase
data['state'] = data['state'].str.strip().str.upper()

# Filter the DataFrame to keep only rows with valid state abbreviations
data = data[data['state'].isin(valid_states)]
data
```

```python
#County-plan pairs that have missing values for eligibles, enrollees, and penetration should be considered as having zero values for these variables.  
data[['eligibles', 'enrollees', 'penetration']] = data[['eligibles', 'enrollees', 'penetration']].fillna(0)
data
```

```python

# Remove duplicate rows
data = data.drop_duplicates()
data
```

```python
# Filter for Puerto Rico and Guam
puerto_rico_guam_data = data[data['state'].str.contains('PR|GU', na=False)]

# Display the filtered data
puerto_rico_guam_data
```

```python
# Exclude Puerto Rico and Guam from the dataset as per instructions
data = data[~data['state'].str.contains('PR|GU', na=False)]

# Display the updated dataset
data
```

```python
#these columns are not needed
data = data.drop(columns=['ABrate', 'monthly_payment'])
data
```

```python
# Filter the data where 'enrollees' is greater than 10
filtered_data = data[data['enrollees'] > 10]

# Group by county and state, then count the number of enrollees
numberofplans1 = filtered_data.groupby(['countyname', 'state']).size().reset_index(name='numberofplans1')
numberofplans1
```

```python
# Filter the data where 'penetration' is greater than 0.5
filtered_data = data[data['penetration'] > 0.5]

# Group by county and state, then count the number of plans
numberofplans2 = filtered_data.groupby(['countyname', 'state']).size().reset_index(name='numberofplans2')
numberofplans2
```

```python

df=data.groupby(['countyname', 'state']).sum().reset_index()
#recreate penetration column
df['penetration'] = df['enrollees'] / df['eligibles']*100


# renaming columns
df.rename(columns={'enrollees': 'totalenrollees', 'penetration': 'totalpenetration'}, inplace=True)
#drop unneeded columns healthplanname typeofplan
df = df.drop(columns=['healthplanname', 'typeofplan'])
df
```

```python
import pandas as pd

# input numberofplans1 and numberofplans2 dataframes
df = df.merge(numberofplans1[['countyname', 'state', 'numberofplans1']], on=['countyname', 'state'], how='left')
df = df.merge(numberofplans2[['countyname', 'state', 'numberofplans2']], on=['countyname', 'state'], how='left')
#fill Nan values with 0 in numberofplans1 and numberofplans2
df[['numberofplans1', 'numberofplans2']] = df[['numberofplans1', 'numberofplans2']].fillna(0)
df
```

```python
def remove_repeated_words(text):
    words = text.split()
    seen = set()
    result = []
    for word in words:
        if word not in seen:
            seen.add(word)
            result.append(word)
    return ' '.join(result)


# Apply the function to the 'countyssa' column
#remove repeated types in SSA column
df['countyssa'] = df['countyssa'].apply(remove_repeated_words)
df
```

```python
#sort by state then county instruction
df = df.sort_values(by=['state', 'countyname']).reset_index(drop=True)

df
# Save the DataFrame to a CSV file
output_file_path = r'C:\Users\clint\Desktop\EvaluativeAssignment_2024\EvaluativeAssignment_2024\Medicare_Advantage\Medicare_Advantage\output_1.csv'
df.to_csv(output_file_path, index=False)
```
